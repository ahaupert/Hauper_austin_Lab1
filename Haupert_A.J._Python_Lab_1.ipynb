```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

S = np.linspace(3,7,9);
W = np.linspace(10,60,6);
E = 0.85

print("FIELD CAPACITY ASSINGMENT")
print()
print("Speed\tWidth\t Capacity")
for x in S:
    for y in W:
        C = x*y*E/8.25;
        print(f'{x}\t{y} \t {C:2.3f} acres/hour')
```
